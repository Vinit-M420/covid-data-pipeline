```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mt
```

```python
top10 = pd.read_csv('../data/top10_owid_covid_loc.csv' ,  usecols = ['location', 'total_cases'])
```

```python
top10.head()
```

```python
top10_plot = top10.set_index('location').plot.bar(ylabel='cases_count',rot=45, figsize = (7,7)) 
```

```python
top10_plot.figure.savefig('top10_total_cases.png')
```

```python
top10death = pd.read_csv('../data/top10_owid_covid_loc.csv' ,  usecols = ['location', 'total_deaths'])
```

```python
top10death
```

```python
top10death.set_index('location').sort_values('total_deaths', ascending=False).plot.bar(ylabel='total_deaths',rot=45, figsize = (7,7)).figure.savefig('top10_total_deaths.png')
```

```python
top10_cd = pd.read_csv('../data/top10_owid_covid_loc.csv' ,  usecols = ['location', 'total_cases', 'total_deaths'])
```

```python
top10_cd['death_rate'] = top10_cd['total_deaths'] * 100/ top10_cd['total_cases']
```

```python
top10_cd.head()
```

```python
sns.set_style('darkgrid')
top10_cd.loc[:, ['location', 'death_rate']].set_index(
    'location').sort_values('death_rate',ascending=False).plot.bar().figure.savefig(
    'top10_deathrate.png')
```
