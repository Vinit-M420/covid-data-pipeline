```python
import pandas as pd
import numpy as np
import matplotlib as mt
import seaborn as sns
```

```python
globalcovid = pd.read_json("../data/global_covid_data.json")
```

```python
globalcovid.head()
```

```python
globalcovid.isna().sum()
```

```python
country_wise_cases = globalcovid.pivot_table(index='country', values=['cases'], aggfunc={'cases':'sum'})
country_wise_cases.head()
```

```python
sns.set_style(style='darkgrid')
country_wise_cases.sort_values('cases' , ascending=False).head().plot.bar(ylabel = 'Cases', xlabel = 'Country')
```
