```python
import requests

response = requests.get("https://disease.sh/v3/covid-19/countries")
data = response.json()
print(data[0])
```

```python
import pandas as pd
import numpy as np
```

```python
covid_data = pd.DataFrame(data)
```

```python
covid_data.tail()
```

```python
covid_data.info(memory_usage = 'deep')
```

```python
covid_data.isna().sum()
```

```python
covid_data.describe().round(2)
```

```python
country_wise_cd = covid_data.pivot_table(index='country', values=['deaths', 'cases'], aggfunc={'deaths':'sum', 'cases':'sum'})
```

```python
country_wise_cd['deaths'].sort_values(ascending = False).head(10)
```

```python
country_wise_cd['cases'].sort_values(ascending = False).head(10)
```
