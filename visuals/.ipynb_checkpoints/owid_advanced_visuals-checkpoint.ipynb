```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
owid_covid = pd.read_csv('../data/owid_covid_cleaned.csv', parse_dates=['date'])
```

```python
owid_covid.head()
```

```python
## Plot Top 10 Countries Over Time
```

```python
owid_covid['year'] = owid_covid['date'].dt.year
owid_covid['year'].unique()
```

```python
owid_covid.pivot_table(index = 'year', values =['new_deaths'], aggfunc='sum')
```

```python
countries_only = owid_covid[owid_covid['iso_code'].str.len() == 3]
```

```python
countries_only.loc[countries_only['iso_code' ]== 'USA'].groupby('year')[['total_deaths']].sum()
```

```python
latest_data = countries_only.sort_values('date').groupby('location', as_index=False).last()
```

```python
top10_countries = latest_data.nlargest(10, 'total_deaths')['location']
```

```python
top10_df = countries_only[countries_only['location'].isin(top10_countries)]
```

```python

sns.lineplot(data=top10_df, x='date', y='total_deaths', hue='location' ).figure.savefig(
    'top10_countriesdeathovertime.png')
```

```python
sns.scatterplot(data = top10_df ,x='total_cases' , y='total_deaths', hue='continent')
```

```python
top10pivot = top10_df.pivot_table(index = 'location', columns=['date'] , values = 'new_cases', aggfunc='sum') 
```

```python
sns.heatmap(data=top10pivot, cmap='Reds', cbar_kws={'label': 'Log(New Cases + 1)'})
```
